```
# Project: Personalized Search using Word Cloud 
#process: Data Wrangling :gather filter convert explore integreate 

#Data gathering 
# a.take user Input
keyword = input("Enter a keyword")

#b. Search keyword on google.com
search_url="https://www.googleapis.com/customsearch/v1/?key=AIzaSyCATL1zbY0i3KDaIKfwz6t6peMyT_vA-0E&cx=012773643409177149611:giuhxzmolrc"
search_request = search_url + "&q="+keyword

#b1.Send a request to googleapis,receieve a response
import requests
response = requests.get(search_request)
print(response.text)

```

```
#Data Filteration and conversion
#Filter jason data from Raw text /Response text
#Filter URl information form natural text
import json 
json_text = json.loads(response.text)

#counter = 1
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#read Url form JSON Data
for item in json_text["items"]:
  title = item["title"]
  url = item["formattedUrl"]
  print(title, "-", url)
  html = item["htmlSnippet"]
  wc = WordCloud().generate(html)
  #obj = "plt" + str(counter)
 # import matplotlib.pyplot as obj
  plt.imshow(wc)
  plt.axis("off")
  plt.show()
  #counter = counter + 1
  
  
```
